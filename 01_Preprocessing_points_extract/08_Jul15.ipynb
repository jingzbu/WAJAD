```python
data_folder = '/home/jzh/Waze/'

import csv
import json

with open(data_folder + 'raw_data/waze_points_Jun15_Dec15.csv', 'rb') as csvfile:
    reader = csv.DictReader(csvfile)

    uuid_list_Jul15 = []
    x_list_Jul15 = []
    y_list_Jul15 = []
    
    for row in reader:
        if '2015-07' in row['\xef\xbb\xbfinject_date']:
            # print(row)
            # assert(1==2)
            # print(row['uuid'], row['x'], row['y'], row['\xef\xbb\xbfinject_date'])
            # print(row['uuid'], row['y'], row['x'])
            uuid_list_Jul15.append(row['uuid'])
            x_list_Jul15.append(row['x'])
            y_list_Jul15.append(row['y'])

# Writing JSON data
with open(data_folder + '/uuid_list_Jul15.json', 'w') as json_file:
    json.dump(uuid_list_Jul15, json_file)

with open(data_folder + '/x_list_Jul15.json', 'w') as json_file:
    json.dump(x_list_Jul15, json_file)
    
with open(data_folder + '/y_list_Jul15.json', 'w') as json_file:
    json.dump(y_list_Jul15, json_file)
```
