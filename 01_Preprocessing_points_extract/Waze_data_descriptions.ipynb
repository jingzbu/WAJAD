## Description of Waze Datasets (Collected in March 2016)

**Jing Zhang** and **Ioannis Ch. Paschalidis** (email: {jzh,yannisp}@bu.edu)

The Waze datasets under investigation are kindly provided to us by the Department of [Innovation and Technology (DoIT) in the City of Boston](https://www.boston.gov/departments/innovation-and-technology). The datasets include three parts: 

(1) the jam data $\mathcal{J}_1$ (traffic slowdown information generated by Waze based on users' location and speed; note that each jam consists of a set of points), 

(2) the corresponding point data $\mathcal{J}_2$ (latitudes and longitudes of the
points within jams), and 

(3) the alert data $\mathcal{J}_3$ (traffic incidents reported by active users; we will call such a user a “Wazer”).

Each entry (jam) in $\mathcal{J}_1$ has the following fields: uuid (unique jam ID), start time, end time, speed (current average speed on jammed segments in meters per second), delay
(delay caused by the jam compared to free flow speed, in seconds), length (jam length in meters), and street (the street where the jam was located). 

The key information for each entry in $\mathcal{J}_2$ includes a jam uuid and the locations (latitudes and
longitudes) of the points within the jam. 

The fields of each entry in $\mathcal{J}_3$ include: uuid (unique system ID; this is different from the jam ID in $\mathcal{J}_1$, because reporting an alert is conducted by a Wazer while recording a jam is done by Waze itself), location (latitude and longitude per report), type (event type; e.g., accident, weather hazard, road closed, etc.), start time, end time, and street (the street where the alert was reported).

The datasets are stored in [JSON](http://www.json.org/) format.

For details of our approach to detecting non-typical Waze jams, please refer to this [paper](https://arxiv.org/abs/1702.08435).
