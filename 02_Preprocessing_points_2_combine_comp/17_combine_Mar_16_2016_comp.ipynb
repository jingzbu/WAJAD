```python
import json
```

```python
# Loading JSON data
with open(data_folder + '/uuid_list_Mar_16_2016.json', 'r') as json_file:
    uuid_list_Mar16 = json.load(json_file)

with open(data_folder + '/x_list_Mar_16_2016.json', 'r') as json_file:
    x_list_Mar16 = json.load(json_file)
    
with open(data_folder + '/y_list_Mar_16_2016.json', 'r') as json_file:
    y_list_Mar16 = json.load(json_file)
```

```python
assert(len(uuid_list_Mar16) == len(x_list_Mar16))
assert(len(uuid_list_Mar16) == len(y_list_Mar16))
```

```python
uuid_list_Mar16[0:30]
```

```python
idx_list = []  # index list for the "turning points;" note that a uuid correspond to many points

L = len(uuid_list_Mar16)  # number of points

for i in range(L-1):
    if uuid_list_Mar16[i+1] != uuid_list_Mar16[i]:
        idx_list.append(i)
#     if len(idx_list) > 2:
#         break    
```

```python
len(idx_list), L
```

```python
import numpy as np

points_dict_Mar16 = {}

x_y_list = []

len_idx_list = len(idx_list)
for i in range(len_idx_list):
    x_y_list = []
    if i == 0:
        x_y_list.append([(float(x_list_Mar16[j]), float(y_list_Mar16[j])) \
                              for j in range(idx_list[i] + 1)])
        numPts = len(range(idx_list[i] + 1))
    else:
        x_y_list.append([(float(x_list_Mar16[j]), float(y_list_Mar16[j])) \
                              for j in range(idx_list[i-1] + 1, idx_list[i] + 1)])
        numPts = len(range(idx_list[i-1] + 1, idx_list[i] + 1))
    data = {"uuid": uuid_list_Mar16[idx_list[i]], "pts": x_y_list, "numPts": numPts}
    points_dict_Mar16['id'+str(i)] = data
#     if i > 3:
#         break
with open(data_folder + '/points_dict_Mar_16_2016_comp.json', 'w') as json_file:
    json.dump(points_dict_Mar16, json_file)
```
