```python
import json
```

```python
# Loading JSON data
with open(data_folder + '/uuid_list_Apr15.json', 'r') as json_file:
    uuid_list_Apr15 = json.load(json_file)

with open(data_folder + '/x_list_Apr15.json', 'r') as json_file:
    x_list_Apr15 = json.load(json_file)
    
with open(data_folder + '/y_list_Apr15.json', 'r') as json_file:
    y_list_Apr15 = json.load(json_file)
```

```python
assert(len(uuid_list_Apr15) == len(x_list_Apr15))
assert(len(uuid_list_Apr15) == len(y_list_Apr15))
```

```python
uuid_list_Apr15[0:30]
```

```python
idx_list = []  # index list for the "turning points;" note that a uuid correspond to many points

L = len(uuid_list_Apr15)  # number of points

for i in range(L-1):
    if uuid_list_Apr15[i+1] != uuid_list_Apr15[i]:
        idx_list.append(i)
#     if len(idx_list) > 2:
#         break    
```

```python
len(idx_list), L
```

```python
import numpy as np

points_dict_Apr15 = {}

x_y_list = []

len_idx_list = len(idx_list)
for i in range(len_idx_list):
    x_y_list = []
    if i == 0:
        x_y_list.append([(float(x_list_Apr15[j]), float(y_list_Apr15[j])) \
                              for j in range(idx_list[i] + 1)])
        x_y_list.append(len(range(idx_list[i] + 1)))
    else:
        x_y_list.append([(float(x_list_Apr15[j]), float(y_list_Apr15[j])) \
                              for j in range(idx_list[i-1] + 1, idx_list[i] + 1)])
        x_y_list.append(len(range(idx_list[i-1] + 1, idx_list[i] + 1)))
    points_dict_Apr15[uuid_list_Apr15[idx_list[i]]] = x_y_list
#     if i > 3:
#         break
with open(data_folder + '/points_dict_Apr15_comp.json', 'w') as json_file:
    json.dump(points_dict_Apr15, json_file)
```
